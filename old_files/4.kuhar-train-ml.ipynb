```python
import os
import json
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
from collections import defaultdict
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix
from typing import Dict, List, Tuple
import matplotlib.pyplot as plt
import yaml

from sklearn.preprocessing import MaxAbsScaler, MinMaxScaler, RobustScaler, StandardScaler
```

# Train Dicrect, time series 300

```python
train_fname = os.path.join("data", "embbedings", "Ku-Har", "train_val_test_300samples", "raw", "1638731242", "train.csv")
val_fname = os.path.join("data", "embbedings", "Ku-Har", "train_val_test_300samples", "raw", "1638731242", "val.csv")
test_fname = os.path.join("data", "embbedings", "Ku-Har", "train_val_test_300samples", "raw", "1638731242", "test.csv")

train = pd.read_csv(train_fname)
val = pd.read_csv(val_fname)
test = pd.read_csv(test_fname)

train = pd.concat([train, val])

X, y = train.iloc[:, 0:1800].to_numpy(), train["class"].to_numpy()
X_test, y_test = test.iloc[:, 0:1800].to_numpy(), test["class"].to_numpy()
print(X.shape, y.shape)
print(X_test.shape, y_test.shape)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import plot_confusion_matrix
model = KNeighborsClassifier(n_neighbors=4)
model.fit(X, y)
y_pred = model.predict(X_test)

print("KNN classifier - ", model)
print(f"F1-score: {f1_score(y_test, y_pred, average='macro'):.4f}")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("Confusion Matrix")
plot_confusion_matrix(model, X_test, y_test)
plt.show()
```

## RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import plot_confusion_matrix
model = RandomForestClassifier(n_estimators=1000)
model.fit(X, y)
y_pred = model.predict(X_test)

print("Random Forest classifier - ", model)
print(f"F1-score: {f1_score(y_test, y_pred, average='macro'):.4f}")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("Confusion Matrix")
plot_confusion_matrix(model, X_test, y_test)
plt.show()
```
